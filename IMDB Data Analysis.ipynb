# 22. Find Unique Values From Genre

```python
Movie_data['Genre']
```

```python
list_genre = []
for values in Movie_data['Genre']:
    list_genre.append(values.split(','))
list_genre
```

```python
final_list = []
for item in list_genre:
    for item1 in item:
        final_list.append(item1)
final_list
    
```

```python
last_list = []
for i in final_list:
    if i not in last_list:
        last_list.append(i)
```
